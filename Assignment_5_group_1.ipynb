```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df_iris = pd.read_csv('C:/Users/MYPC/Downloads/archive (2)/Iris.csv')


# Define color mapping for species
species_colors = {"Iris-setosa": "blue", "Iris-versicolor": "green", "Iris-virginica": "red"}

# Improved visualization
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x=df_iris['SepalLengthCm'],
    y=df_iris['SepalWidthCm'],
    hue=df_iris['Species'],
    palette=species_colors,
    style=df_iris['Species'],
    s=100  # Adjust marker size
)

# Adding labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs. Sepal Width by Species")
plt.legend(title="Species")

# Display the improved plot
plt.show()
```

**Improvements Made:**


**Color Coding by Species:** Different colors are used to distinguish between Iris-setosa, Iris-versicolor, and Iris-virginica. This helps in understanding patterns and distribution across species.


**Marker Styles for Better Differentiation:** Unique markers (circles, crosses, and squares) help differentiate species even in black-and-white prints.



**Larger Figure Size:** The visualization is now larger, making it easier to read.



**Clear Axis Labels:** Labels for Sepal Length and Sepal Width are added, improving readability.



**Title for Context:** A more descriptive title provides clarity.



**Legend for Quick Identification:** The legend clarifies the mapping of colors and styles to species.



These enhancements make the visualization more informative and accessible, helping to identify patterns in the dataset more effectively.

**Question 2: Amazon prime Title dataset**

https://www.kaggle.com/datasets/shivamb/amazon-prime-movies-and-tv-shows

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

df_amazon = pd.read_csv('/content/amazon_prime_titles.csv')

rating_counts = df_amazon['rating'].value_counts()
```

```python
# Bad Visualization
plt.figure(figsize=(5, 5))
df_amazon['rating'].value_counts().plot(kind='bar', color='purple')
plt.title('Content Rating Distribution')
plt.show()
```

Task 2: Improve the Visualization

Create a bar chart that addresses the above issues.
Add appropriate axis labels and a title.
Display the count of each rating category on the bars.
Choose a color palette that enhances readability and accessibility.

Provide a Brief Explanation:

Describe the improvements made and how they enhance the visualization's effectiveness.

```python
# Improved Viz
plt.figure(figsize=(10, 6))
bars = plt.bar(rating_counts.index, rating_counts.values, color=['#4C72B0', '#DD8452', '#55A868', '#C44E52', '#8172B2'])

# Adding count labels on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, str(bar.get_height()), ha='center', fontsize=7, fontweight='bold')

# Titles and labels
plt.title('Distribution of Content Ratings on Amazon Prime', fontsize=14, fontweight='bold')
plt.xlabel('Rating Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotating labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding a light grid

plt.show()
```

### **Improvements**
1. **Increased Figure Size:**  
   - The graph is now larger (`10x6`), making it easier to read and interpret.

2. **More Readable Color Palette:**  
   - Used distinct colors from an accessibility-friendly palette for better visibility.

3. **Count Labels on Bars:**  
   - Each bar displays the exact count, making it easier to interpret values without guessing.

4. **Better Title & Labels:**  
   - The title and axis labels are clearer, making the graph more informative.

5. **Rotated X-axis Labels:**  
   - Prevents overlapping and improves readability, especially for long rating categories.

6. **Grid Lines for Better Interpretation:**  
   - Light horizontal grid lines make it easier to compare bar heights visually.
